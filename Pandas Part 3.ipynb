```python
## Lets talk about a two different differnet operation which is called as Merge and join operation
```

```python
data1 = pd.DataFrame({'key1':[2,3,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[2,3,4,5,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)                     
```

```python
data1
```

```python
data2
```

```python
## perform merge operation
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2,how = 'left')
```

```python
data1
```

```python
pd.merge(data1,data2,how = 'right')
```

```python
pd.merge(data1,data2,how = 'outer')
```

```python
## cross means try to create a combination of all the possible outcome that is called cross merge opertion
pd.merge(data1,data2,how = 'cross')
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,how = 'left' ,left_on = "key2",right_on = 'key4')
```
